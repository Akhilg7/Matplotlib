```python
import matplotlib.pyplot as plt
import numpy as np
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
Month=np.arange(1,13)
plt.plot(Month,Max,'ro')
plt.plot(Month,Min,'bo')
plt.xlabel("Month")
plt.ylabel("Temparature")
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, Month,
                                      Max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, Month,
                                      Min, [-40, 20, 0])
print(res_max,cov_max)
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(Month, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(Month, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
1. Create a pie chart presenting the male/female proportion 
```

```python
import pandas as pd
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic=pd.read_csv(url)
titanic
```

```python
y=[0,0]
for i in titanic['sex']:
    if i=='male':
        y[0]+=1
    if i=='female':
        y[1]+=1
```

```python
import matplotlib.pyplot as plt
import numpy as np
label=['Male','Female']
plt.pie(y,labels=label,autopct='%1.2f%%')
```

```python
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
```

```python
#Approach 1
import numpy as np
import matplotlib.pyplot as plt
mage=[]
mfare=[]
fage=[]
ffare=[]

for i in range(len(titanic)):
    if titanic['sex'][i]=='male':
        mage.append(titanic['age'][i])
        mfare.append(titanic['fare'][i])
        
    if titanic['sex'][i]=='female':
        fage.append(titanic['age'][i])
        ffare.append(titanic['fare'][i])
plt.scatter(mage,mfare,c='r')
plt.scatter(fage,ffare,c='g')
plt.xlabel('age')
plt.ylabel('fare')
plt.legend(['male','female'])
```
